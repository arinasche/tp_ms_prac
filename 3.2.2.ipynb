```python
from pandas.plotting import parallel_coordinates
import seaborn as sb
titanic=sb.load_dataset("titanic")
plt.figure(figsize=(10, 10)) 
tianic=titanic.drop(['sex', 'sibsp','parch','embarked','adult_male','deck','embark_town','alone','class'], axis='columns', inplace=True)
titanic=titanic.replace({'who':{'man': '1', 'woman': '2', 'child':'3'}})
titanic[["who"]] = titanic[["who"]].apply(pd.to_numeric)
parallel_coordinates(titanic, 'alive', colormap='autumn')  
```
