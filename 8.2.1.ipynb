```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import math
n=5
a=0.05
c=1-(a**(1/n))
print(c)
plt.figure(figsize=(10,5))
def p1(x):
    if (x<=c) and (c<=x+1):
        return((1+x-c)**n)
    elif  c<x:
        return(1)
    else:
        return(0)
z=np.vectorize(p1,otypes=[np.float])
x = np.linspace(-5, 5 , 1000)
plt.plot(x,z(x),label = 'мощность')
plt.legend()
plt.title('$X_{(1)}>C$')
plt.show()
```

```python
plt.figure(figsize=(10,5))
def p2(x):
    if (x<=c) and (c<=x+1):
        return(1-(1+x-c)**n)
    elif  c<x:
        return(0)
    else:
        return(1)
z=np.vectorize(p2,otypes=[np.float])
x = np.linspace(-5, 5 , 1000)
plt.plot(x,z(x),label = 'мощность')
plt.legend()
plt.title('$X_{(1)}<C$')
plt.show()
```

```python
plt.figure(figsize=(10,5))
def p2(x1):
    if (x1<=c) and (c<=x1+1):
        return(1-(-x1+c)**n)
    return(0)
z=np.vectorize(p1,otypes=[np.float])
x = np.linspace(-5, 5 , 1000)
plt.plot(x,z(x),label = 'мощность')
plt.legend()
plt.title('$X_{(n)}>C$')
plt.show()
```
