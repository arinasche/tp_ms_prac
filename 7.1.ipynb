```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import brewer2mpl
from matplotlib import rcParams
import seaborn as sb
import scipy.stats as sc
a=[0.001,0.05,0.1]
for i in a:
    x = np.linspace(0, i, 10000)
    plt.figure(figsize=(10, 5))
    plt.plot(x, sc.norm.ppf(1-i+x)-sc.norm.ppf(x))
    xlabel('b')
    ylabel('y')
    title(i)
    show()
```

```python
m=[1,2,5,10,100]
for i in a:
    for n in m:
        x = np.linspace(0, i, 10000)
        s='n='+str(n)
        plt.figure(figsize=(8, 4))
        plt.plot(x, sc.gamma.ppf(1-i+x,n,1)-sc.gamma.ppf(x,n,1),label = s)
        legend()
        xlabel('b')
        ylabel('y')
        title(i)
        show()
```
