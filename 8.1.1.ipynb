```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import math
n=10
plt.figure(figsize=(10,5))
def p1(x):
    if (0<=x) and (x<=1):
        return((1-x)**n)
    elif  x<0:
        return(1)
    else:
        return(0)
def p2(x1):
    if (1<x1) and (x1<=2):
        return(1-(2-x1)**n)
    return(0)
x = np.linspace(0, 2 , 1000)
y=np.vectorize(p1,otypes=[np.float])
z=np.vectorize(p2,otypes=[np.float])
plt.plot(x,y(x),label = '1')
plt.plot(x,z(x),label = r'${2, \theta_1=1}$ ')
plt.legend()
plt.title('$X_{(1)}>C$')
plt.show()
```
