### Задание 1

```python
n = 1000
lastzero = []
for i in range(2000):
    s = []
    s.append(0)
    for j in range(1,n+1):
        s.append(s[j-1]+np.random.choice([-1,1]))
    lastzero.append(len(s)-1-list(reversed(s)).index(0))
plt.figure(figsize = (5,4))
x = np.linspace(0, 1 , 2000)

plt.plot(x, np.cumsum(lastzero) / sum(lastzero), label = 'ЭФР')
plt.plot(x, sc.arcsine.cdf(x), label = 'arcsin')
plt.legend()
```

Из Лагутина:

$$ E\nu = \frac{-ar_a+(B-a)(1-r_a)}{2p-1}$$,
где $r_a = \frac{\lambda^B - \lambda^a}{\lambda^B - 1}, \lambda = \frac{p}{q}$

```python
n = 100
P = [0.9, 0.5, 0.1]
a = 100
B = [1.5 * a, 2 * a, 10 * a]

for p in P:
    for b in B:
        l = (1 - p) / p
        if 2 * p  - 1 and p - 0.1 :
            r = (l ** b - l ** a)/(l ** b - 1)
            warnings.filterwarnings("ignore")
        res = 0
        t = []
        for i in range(n):
            s = a
            m = 0
            while (s>0) & (s<b):
                s += np.random.choice([-1, 1], p = [1 - p, p] )
                m += 1  
            res += int(s > 0)
            t.append(m) 
        print( '\n Для p =' + str(p) + ' и b=' + str(b)+' эмпирическая вероятность:' + str(res / n))
        print('Среднее эмпирическое время остановки: ' + str(np.mean(t)))
        if (2 * p -1) and ((b != B[2]) or (p != P[2])):
            print('Среднее теоретическое время остановки: ' + str((-a*r +(b-a)*(1-r))/(2*p-1)))
            warnings.filterwarnings("ignore")
```
