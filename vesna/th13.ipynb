```python
import matplotlib.pyplot as plt
import scipy.stats as sc
import seaborn as sb
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
N = 50
K = 5
```

```python
a_first = np.zeros(K)
a_sec = np.zeros(K)
a_sec[K-1] = 0.5
a_th = [i / 10 for i in range(1,K+1)]
K
```

### Для $\mathcal{N}(a_i, 1)$

```python
p_f = []
p_k = []
for n in range(1,N+1):
    sample = [[] for i in range(K)]
    for i in range(K):
        sample[i] = sc.norm.rvs(loc = a_first[i], size = n)
    p_f.append(sc.f_oneway(sample[0], sample[1], sample[2], sample[3], sample[4])[1])
    p_k.append(sc.kruskal(sample[0], sample[1], sample[2], sample[3], sample[4])[1])
p_f = np.sort(p_f)
p_k = np.sort(p_k)
v = np.arange(0, 1, 1/N)
plt.figure(figsize = (5,4))
plt.plot(p_f, v, label ='F-тест')
plt.plot(p_k, v, label ='К-У')
plt.legend()
plt.show()
```
