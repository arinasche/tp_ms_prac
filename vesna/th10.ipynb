### Проиллюстрировать графически для процесса восстановления ν(t) утверждение:

$\frac{\nu(t)}{t} \rightarrow \frac{1}{a}$

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sc
import warnings
```

```python
n = 3000
m = 3000
l = 20
ar_t = sc.expon.rvs(scale = 1/l, size = n)
arrival_times = []
for i in range(n):
    if i == 0:
        arrival_times.append(ar_t[i])
    else:
        arrival_times.append(ar_t[i]+arrival_times[i-1])
arrival_times = np.array(arrival_times)
f = lambda x: len(arrival_times[arrival_times < x])
plt.figure(figsize = (6,5))
t = np.linspace(1,max(arrival_times+1),m)
a = [ f(t[i]) / t[i] for i in range(len(t))]
b = [l]*m
plt.plot(t,a)
plt.plot(t,b)
plt.show()
```

### Задача 2

```python
def f_pose(t,b,n):
    for i in range(n):
        if t < b[i]:
            return b[i] - t
```

```python
n = 3000
m = 3000
l = 1
ar_t = sc.expon.rvs(scale = 1/l, size = n)
arrival_times = []
for i in range(n):
    if i == 0:
        arrival_times.append(ar_t[i])
    else:
        arrival_times.append(ar_t[i]+arrival_times[i-1])
arrival_times = np.array(arrival_times)
t = sc.uniform.rvs(size = m)
exc =[f_pose(t[i]*max(arrival_times),arrival_times,n) for i in range(len(t))]
print(f'Оценка математического ожидания: {np.mean(exc)}')
plt.figure(figsize = (6,5))
sb.histplot(exc, kde = True)
plt.show()
```
