```python
import pandas as pd
from scipy.stats import chisquare 
import numpy as np
import scipy.stats as sc 
df = pd.read_csv('Priem_utf.csv', sep=';' )
```

```python
s=df[df	["ГТО.оценка"] == 0]
b=df[df	["ГТО.оценка"] != 0]
s1=s.loc[:, 'Сумма.баллов'].to_numpy()
b1=b.loc[:, 'Сумма.баллов'].to_numpy()
ch1=[]
ch2=[]
for i in range(196,408,10):
    cond_1=np.logical_and(s1>=i,s1<i+10)
    cond_2=np.logical_and(b1>=i,b1<i+10)
    ch1.append(len(s1[cond_1]))
    ch2.append(len(b1[cond_2]))
q=sc.chi2_contingency([ch1,ch2])
print("Значение статистики: "+str(q[0]))
print("p-value: "+str(q[1]))
print("Ожидаемые: "+str(q[3]))
```
