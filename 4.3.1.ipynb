```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as sc
m=[2,5,10,20,50,100,500]
for n in m:
    s=[]
    for i in range(0,1000):
        x=np.random.exponential(scale=1.0, size=n)
        s.append(min(x))
    plt.figure(figsize = (5,4))
    sb.distplot(s,fit = sc.norm, label = '$X_{1}$') 
```

```python
m=[2,5,10,20,50,100,500]
for n in m:
    s=[]
    for i in range(0,1000):
        x=np.random.exponential(scale=1.0, size=n)
        s.append(n*(min(x+2)-2))
    plt.figure(figsize = (5,4))
    sb.distplot(s,fit = sc.expon, ) 
```

```python
m=[2,5,10,20,50,100,500]
for n in m:
    s=[]
    t=[]
    for i in range(0,1000):
        x=np.random.exponential(scale=1.0, size=n)
        s.append(min(x+2)-1/n)
        t.append(np.mean(x+2)-1)
    plt.figure(figsize = (8,6))
    sb.distplot(s, fit=sc.expon, label = 'X(1)') 
    sb.distplot(t,  label = 'выб.ср.-1') 
    plt.legend()
    plt.show()
```

```python
m=[2,5,10,20,50,100,500]
from math import sqrt
for n in m:
    s=[]
    t=[]
    for i in range(0,1000):
        x=np.random.exponential(scale=1.0, size=n)
        s.append(sqrt(n)*(np.mean(x+2)-1-2))
        t.append(sqrt(n)*(sqrt(np.mean((x+2)**2-2*(x+2)))-2))
    plt.figure(figsize = (8,6))
    sb.distplot(s,  label = '1') 
    sb.distplot(t,  label = '2') 
    plt.legend()
    plt.show()
```
