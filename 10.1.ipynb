```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
k=[1,2,10]
n=[10, 50, 200]
m=[100,500]
for l in range(len(n)):
    for j in range(len(k)):
        for  i in range(len(m)):
            p=[]
            for s in range(m[i]):
                x=[]
                for h in range(n[l]):
                        y = sc.uniform.rvs(size = k[j])
                        x.append((np.sum(y)-(k[j]/2))/np.sqrt(k[j]/12))
                p.append(sc.kstest(x,sc.norm.cdf).pvalue)
            p.sort()
            x1 = np.arange(0,1,1/m[i])
            plt.figure( figsize= (8,6))
            plt.plot(list(map(lambda x1: p[int(x1*m[i])], x1)),x1 ,label = ("n = "+str(n[l])+"; k = "+str(k[j])+"; m = "+str(m[i])))
            plt.xlim(0,1)
            plt.legend()
```
